# Categorise the states into four categories based on GDP per capita

```python
#PerCapita.loc[0]<PerCapita.quantile(.85,axis=1)
#PerCapita.loc[:,(PerCapita.loc[0] < (PerCapita.quantile(.85,axis=1)))]
#PerCapita.loc[:,(PerCapita.loc[0] > 153246)]
#PerCapita.iloc[0] < 62091

q1=PerCapita.quantile(.85,axis=1)   #Assign the quantile values to q1
C1=PerCapita.loc[:,(PerCapita.loc[0] > q1[0])]
C1
```

```python
q1=PerCapita.quantile(.85,axis=1)
q2=PerCapita.quantile(.5,axis=1)
C2=PerCapita.loc[:,((PerCapita.loc[0] < q1[0]) & (PerCapita.loc[0]>q2[0]))]
C2
```

```python
q2=PerCapita.quantile(.5,axis=1)
q3=PerCapita.quantile(.2,axis=1)
C3=PerCapita.loc[:,((PerCapita.loc[0] < q2[0]) & (PerCapita.loc[0]>q3[0]))]
C3
```

```python
q3=PerCapita.quantile(.2,axis=1)
C4=PerCapita.loc[:,(PerCapita.loc[0] < q3[0])]
C4
```

```python
# Function to select only the sub sectors

def sector(n):
    sec=n
    df3= pd.DataFrame()
    new1=pd.DataFrame()
    df4=pd.DataFrame()
    for state in sec:
    
        C1df=Data1B.loc[["Agriculture, forestry and fishing","Mining and quarrying","Manufacturing","Electricity, gas, water supply & other utility services","Construction","Trade, repair, hotels and restaurants","Transport, storage, communication & services related to broadcasting","Financial services","Real estate, ownership of dwelling & professional services","Public administration","Other services","Gross State Domestic Product"],state]
    #dff.columns=[state]
        new1=C1df
        df3=pd.concat([df4,new1], axis = 1)  #Concat the dfs to single df
        df4=df3
    
    return(df4)
   
    
```

```python
C1_Sector=sector(C1)
C2_Sector=sector(C2)
C3_Sector=sector(C3)
C4_Sector=sector(C4)
```

```python
C1_Sector
```

```python
#Take the sum of all the sub sectors
C1_Sector["C1_Sum"]=C1_Sector.sum(axis=1)
C2_Sector["C2_Sum"]=C2_Sector.sum(axis=1)
C3_Sector["C3_Sum"]=C3_Sector.sum(axis=1)
C4_Sector["C4_Sum"]=C4_Sector.sum(axis=1)
```

```python
C4_Sector["C4_Sum"]
```

```python
#Find the percentage of sub sectors to the GDP

C1_SectorPercent=C1_Sector.loc[:,["C1_Sum"]]/C1_Sector.loc["Gross State Domestic Product",["C1_Sum"]]*100
C2_SectorPercent=C2_Sector.loc[:,["C2_Sum"]]/C2_Sector.loc["Gross State Domestic Product",["C2_Sum"]]*100
C3_SectorPercent=C3_Sector.loc[:,["C3_Sum"]]/C3_Sector.loc["Gross State Domestic Product",["C3_Sum"]]*100
C4_SectorPercent=C4_Sector.loc[:,["C4_Sum"]]/C4_Sector.loc["Gross State Domestic Product",["C4_Sum"]]*100
```

```python
C4_SectorPercent
```

```python
C1_SectorPercent.sort_values(by="C1_Sum",ascending=False,inplace=True)
C2_SectorPercent.sort_values(by="C2_Sum",ascending=False,inplace=True)
C3_SectorPercent.sort_values(by="C3_Sum",ascending=False,inplace=True)
C4_SectorPercent.sort_values(by="C4_Sum",ascending=False,inplace=True)
```

```python
C2_SectorPercent
```

```python
#Remove GDP row as not needed
C1_SectorPercent=C1_SectorPercent[1:]
C2_SectorPercent=C2_SectorPercent[1:]
C3_SectorPercent=C3_SectorPercent[1:]
C4_SectorPercent=C4_SectorPercent[1:]
```

```python
#Concat all the dfs to "subSectors"
subSectors=pd.concat([C1_SectorPercent,C2_SectorPercent,C3_SectorPercent,C4_SectorPercent],axis=1,sort=True)
```

```python
#Transpose df
subSectors=subSectors.T
```

```python
subSectors.sort_values(by=["C2_Sum","C1_Sum","C3_Sum","C4_Sum"],axis=1,ascending=False,inplace=True)
```

```python
subSectors
```

```python
#Plot the contribution of the sub-sectors as a percentage of the GSDP of each category. 

subSectors.plot(kind="bar",stacked="True",figsize=(20,10),legend=False)
plt.ylabel("GDP % Contribution")
leg = plt.legend( loc = 'upper right')

```

```python
plt.figure(figsize=(10,5))
sns.heatmap(subSectors, cmap="YlGnBu", annot=True)

#plt.xticks(rotation=85)
plt.show()
```

# Dropout rates divided into Categegories

```python
Dropouts_GDP.set_index("State",inplace=True)
```

```python
def categoryDropouts(c):
    drop=c
    df5= pd.DataFrame()
    new2=pd.DataFrame()
    df6=pd.DataFrame()
    for state in drop:
        
        state=state.replace('_',' ')
        if state=='Uttarakhand':
            state='Uttrakhand'
        elif state=='Chhattisgarh':
            continue
        D1df=Dropouts_GDP.loc[[state]]
        new2=D1df
        df5=pd.concat([df6,new2], axis = 0)  #Concat the dfs to single df
        df6=df5
    
    return(df6)
```
